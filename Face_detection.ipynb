```python
import cv2
from matplotlib import pyplot as pyplot
from google.colab import files
```

```python
image_file = '/content/56fec125a824d4a1c4430abe9a1ef3b1.png'
image = cv2.imread(image_file)
```

```python
image = cv2.imread(image_file)
image_rgb = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
```

```python
!wget -O haar_face.xml https://github.com/opencv/opencv/raw/master/data/haarcascades/haarcascade_frontalface_default.xml
face_cascade = cv2.CascadeClassifier('haar_face.xml')
```

```python
gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
```

```python
faces = face_cascade.detectMultiScale(gray,scaleFactor = 1.1 ,minNeighbors=5)
```

```python
for(x,y,w,h) in faces:
  cv2.rectangle(image_rgb,(x,y),(x+w ,y+h),(255,0,0),2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.imshow(image_rgb)
plt.title(f'Faces datected:{len(faces)}')
plt.axis('off')
plt.show()
```
